```python
import pandas as pd
import numpy as np 
import scipy.stats as st

veri1 = [4]*2+[6]*3+[8]*50+[10]*130+[12]*220+[14]*402+[16]*508+[18]*440+[20]*180+[22]*90+[24]*70+[26]*4+[28]

unique, counts = np.unique(veri1, return_counts = True)
X = np.asarray((unique))
G = np.asarray((counts)) 

tablo1 = pd.DataFrame({"Xj":X, "fj":G})
tablo1["Xjfj"] = X * G
n = G.sum()
xf = np.sum(X*G)
m = xf/n # ortalama 15.92
 
tablo1["Xj-X"] = X - np.round(m,2)
tablo1["(Xj-X)^2"] = np.round(np.square(X - m),4)
tablo1["(Xj-X)^2fj"] = np.round(np.square(X - m),4)*G
gt = tablo1["(Xj-X)^2fj"].sum()
v = (1 / (n - 1))* gt #varyans 12.62
tablo1
```

### Standartlaştırma

```python
tablo2 = pd.DataFrame({"Gj" : G})
alt = np.arange(3,28,2)
ust = np.arange(5,30,2)
za,zu,p,b = [],[],[],[]

#alt ve üst değerlerin standartlaştırılması
for i in range(len(tablo2.index)):
    za.append((alt[i] - m) / v**(1/2))
    zu.append((ust[i] - m) / v**(1/2))
za = np.round(za,2)
zu = np.round(zu,2)

tablo2["Zaj"] = za
tablo2["Zuj"] = zu

for i in range(len(tablo2.index)):
    p.append(np.round(st.norm.cdf(zu[i]) - st.norm.cdf(za[i]), 4)) 
    
#Pj ve beklenen değerlerin bulunması

p[0] = np.round(1 - st.norm.cdf(3.08),4)
p[-1] = np.round(st.norm.cdf(3.12) - 1,4) # olasılıklar toplamı 1 olacak

for i in range(len(tablo2.index)):
    b.append(p[i] * n) # beklenen frekanslar
B = np.abs(b)
p = np.abs(p)
tablo2["Pj"], tablo2["Bj"] = p, B
tablo2
```

### Test İstatistiğinin Hesaplanması

```python
tablo3 = pd.DataFrame({"Gj" : G, "Bj":B})
ist = []
for i in range(len(tablo3.index)):
    ist.append(((G[i] - B[i])**2) / B[i])
tablo3["(Gj-Bj)^2/Bj"] = np.round(ist, 3)
tablo3
```

### Test İstatistiği

```python
xh = tablo3["(Gj-Bj)^2/Bj"].sum()
xh
```

### Kritik değer

```python
a = 0.05
xt = st.chi2.ppf(1-(a), 10)
xt
```

```python
if xh > xt:
    print("Ho hipotezi reddedilir")
else:
    print("Ho hipotezi reddedilemez")
```
